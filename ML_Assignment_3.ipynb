```python
# Q9 - Display the percentage of working men and women in the dataframe?

g = len(df.gender)
m = len(df[df['gender'] == 'Male'])

f = len(df[df['gender'] == 'Female'])

print("% of working men",100*(m/g))

print("% of working women",100*(f/g))
```

```python
# Q12.2 - Display the basic statistical analysis of 'hours-per-week' from male and female point


m = df["hours-per-week"].loc[df.gender == "Male"].sum()
print("Total hours of male : ",m)

m = df["hours-per-week"].loc[df.gender == "Male"].describe()
print(m)

f = df["hours-per-week"].loc[df.gender == "Female"].sum()
print("Total hours of female :",f)

f = df["hours-per-week"].loc[df.gender == "Female"].describe()
print(f)
```

```python
# Q13.1 - Groupby on the race column and display first()

race = df.groupby('race')
race.first()
```

```python
# Q13.2 - Display the basic statistical analysis of 'hours-per-week' for different race


Black = df["hours-per-week"].loc[df.race == "Black"].describe()
print("Black:\n", Black,'\n')

White = df["hours-per-week"].loc[df.race == "White"].describe()
print("White:\n",White, "\n")

Amer_Indian_Eskimo = df["hours-per-week"].loc[df.race == "Amer-Indian-Eskimo"].describe()
print("Amer-Indian-Eskimo:\n",Amer_Indian_Eskimo, "\n")

Asian_Pac_Islander = df["hours-per-week"].loc[df.race == "Asian-Pac-Islander"].describe()
print("Asian-Pac-Islander:\n",Asian_Pac_Islander, "\n")

Other = df["hours-per-week"].loc[df.race == "Other"].describe()
print("Other:\n",Other)
```
